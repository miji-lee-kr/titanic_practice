```python
# object comparison

x = ['orange', 'banana', 'apple'] # 원본
y = x.copy() # shallow copy
z = ['orange', 'banana', 'apple'] # 새 리스트

print(x == y) # 원본과 값 동일
print(x == z) # 원본과 값 동일

print(x in y) # 복사 했어도 객체가 다름
# 동일하게 유지되는 것은 리스트 내 데이터 (e.g.orange)뿐이고, 리스트 틀[]은 새로 만들어짐
# 이 때 리스트 안의 값이 가변 타입 (e.g. 중첩 리스트, 중첩 딕셔너리, 중첩 집합)이면 틀뿐만 아니라 내부 데이터까지 바뀜

print(x in z) # 새로 만들었으니 당연히 객체 상이
```

```python
x = ["grapes", "mango", "orange", "peach", "apple", "lime", "banana", "cherry", "tomato", "kiwi", "blueberry", "watermelon"]

# 출력 결과 : [APPLE, KIWI] 로 만들기

# 방법 1.

emp = []
for value in x: 
    if value == 'apple' or value=='kiwi':
        new_val = value.upper()
        emp.append(new_val)
print(emp)


# 방법 2. x 문자열 자체를 가져오기보다 for i in range(len(x)) 로 리스트를 가져오는게 활용도가 높을 듯

emp = []
for i in range(len(x)): 
    if x[i] == 'apple' or x[i] == 'kiwi':
        emp.append(x[i].upper())
print(emp)


# 방법 3. filter() 사용 - filter(func, iterable 객체)

# step 1. filter함수는 리스트 내 개별값에 접근할 수 있다. 개별값이 참(apple or kiwi) 인지 아닌지 거르자 (개별값에 접근하면 for로 한바퀴 돌 필요 없음)
# step 2. upper() 적용하자. 
# 이때 upper() 는 문자열에만 적용할 수 있음. 리스트에 못씀. filter 객체에도 못씀
# 'apple'.upper()는 돼도 ['apple'].upper() 안됨. (x.apply(filter(lambda x: x =='apple' or x == 'kiwi', x)불가)
# -> 그러니 filter처럼 개별값에 접근해서 함수로 바꾸자

list(map(lambda y: y.upper(), filter(lambda x: x =='apple' or x == 'kiwi', x)))


# 방법 4. list comprehension 만 사용
# for 문 역시 리스트 개별 원소에 접근해서 kiwi인지 구분할 수 있고, upper() 적용할 수 있음
# 이 for문을 더 간단하게 써보자 

z = [x[i] for i in range(len(x)) if x[i] == 'kiwi' or x[i] == 'apple']
list(i.upper() for i in z)
```

```python
word = 'secrect' # 정답 단어
guesses = '' # 추측 단어
turns = 10 # 기회 10번

# 기회 1번 남을 때까지 계속 함
while turns >0:
    failed = 0 # 실패 횟수
    
    # 정답 맞추게 제시.
    # 단어에 있으면 출력, 없으면 failed 에 추가
    for char in word:
        if char in guesses:
            print(char, end = ' ')
        else:
            print("_", end = ' ')
            failed += 1
            
    # 단어 다 맞췃으면 while 구문 중단  
    if failed == 0:
        print("Correct")
        break
    guess = input("your turn")
    guesses += guess
    
    # 정답 단어에 추측 문자 없으면
    if guess not in word:
        turns -= 1
        print("Wrong. You have", turns, 'more guess')
        
    # 기회 다 쓰면 게임 오버
    if turns == 0:
        print('Game over')
```
