```python
# object comparison

x = ['orange', 'banana', 'apple'] # 원본
y = x.copy() # shallow copy
z = ['orange', 'banana', 'apple'] # 새 리스트

print(x == y) # 원본과 값 동일
print(x == z) # 원본과 값 동일

print(x in y) # 복사 했어도 객체가 다름
# 동일하게 유지되는 것은 리스트 내 데이터 (e.g.orange)뿐이고, 리스트 틀[]은 새로 만들어짐
# 이 때 리스트 안의 값이 가변 타입 (e.g. 중첩 리스트, 중첩 딕셔너리, 중첩 집합)이면 틀뿐만 아니라 내부 데이터까지 바뀜

print(x in z) # 새로 만들었으니 당연히 객체 상이
```

```python
x = ["grapes", "mango", "orange", "peach", "apple", "lime", "banana", "cherry", "tomato", "kiwi", "blueberry", "watermelon"]

# 출력 결과 : [APPLE, KIWI] 로 만들기

# 방법 1.

emp = []
for value in x: 
    if value == 'apple' or value=='kiwi':
        new_val = value.upper()
        emp.append(new_val)
print(emp)


# 방법 2. x 문자열 자체를 가져오기보다 for i in range(len(x)) 로 리스트를 가져오는게 활용도가 높을 듯

emp = []
for i in range(len(x)): 
    if x[i] == 'apple' or x[i] == 'kiwi':
        emp.append(x[i].upper())
print(emp)


# 방법 3. filter() 사용 - filter(func, iterable 객체)

# step 1. filter함수는 리스트 내 개별값에 접근할 수 있다. 개별값이 참(apple or kiwi) 인지 아닌지 거르자 (개별값에 접근하면 for로 한바퀴 돌 필요 없음)
# step 2. upper() 적용하자. 
# 이때 upper() 는 문자열에만 적용할 수 있음. 리스트에 못씀. filter 객체에도 못씀
# 'apple'.upper()는 돼도 ['apple'].upper() 안됨. (x.apply(filter(lambda x: x =='apple' or x == 'kiwi', x)불가)
# -> 그러니 filter처럼 개별값에 접근해서 함수로 바꾸자

list(map(lambda y: y.upper(), filter(lambda x: x =='apple' or x == 'kiwi', x)))


# 방법 4. list comprehension 만 사용
# for 문 역시 리스트 개별 원소에 접근해서 kiwi인지 구분할 수 있고, upper() 적용할 수 있음
# 이 for문을 더 간단하게 써보자 

z = [x[i] for i in range(len(x)) if x[i] == 'kiwi' or x[i] == 'apple']
list(i.upper() for i in z)
```

```python
# Dict value들의 합 구하기 - 복습


d = {'a': 17,'b':8, 'c':20}


# 1. # 무조건 리스트에 append로 담기만 하지 말고 어차피 숫자 합이니까 합을 업데이트 하기   
summed = 0
for v in d:
    summed += d.get(v)
print(summed)


# 2.
summed = 0
for value in d.values():
    summed += value
print(summed) 


# 3. 
sum([d[key] for key in d]) # dict 전체에서 어차피 key값만 도니까 d[key]로 value 빼냄  

```

```python
# value 값이 25 이상인 Dict 요소만 골라내자 - 복습

d = {'a': 8, 'b':33, 'c': 3}

# 1. 일반 루프

emp = {}

for key, value in d.items(): # 그냥 for k,v in d: 로 하면 안됨.
                        # d는 dict 전체를 받았어도 루프를 도는건 key뿐이다. value는 print(value) 처럼 아래에서 따로 부르지 않는 한 숨어있음
                        # d.items()는 (a,8) 같이 key, value 모두 있는 하나의 튜플을 받고, 튜플 각 쌍이 iterate한다.
                        # 즉 둘 다 필요하다면 for in d.items()를 써야함. value만 필요하면 for in d.value()를 받자.
                        # 한번 돌아서 받은 (a, 8) 튜플은 for구문이 자동으로 언패킹해서 a는 key에, 8은 value에 자동으로 어사인해준다 = 언패킹.
                        # d는 key라는 하나의 변수밖에 못받으므로 key, value 두개 요청한 갯수랑 안맞아 에러 
    if value >=25:
        emp[key] = value
        # emp에 append 안하고 if value >=25: print(d) 하니 모든 딕셔너리가 나왔던 실수 -
        # print()에 전달된건 여전히 d 전체이므로, value >=25 인 값이 하나라도 있으면 전체를 표시하라는 지시가 됨
        # 차라리 print(key,value)하면 해당되는 부분만 나옴
print(emp)


# 2. lambda

dict_lambda = dict(filter(lambda x: x[1] >=25, d.items())) # d만 쓰면 lambda도 for처럼 key만 받음. 람다에서도 d.items() 를 받아야 
                                                            # d.items() 로 한번 루프 돌 때마다 (a,8) 같이 두개 변수를 가진 한 덩이의 튜플 생김
                                                            # lambda는 하나의 변수만 받아줌 & for처럼 key, value for d.items() 같은 자동 언팩킹 안해줌 -> key, value에 자동으로 값 assign 안해줌
                                                            # 우리가 튜플 안에서 뭐가 value인지 수동으로 assign 해줘야함 -> (a,8)에서 1번째 인덱스가 value = tuple[1]
                                                            # key>=25인 튜플 덩어리만 필터링해서 그 덩어리를 dict로 다시 패킹
print(dict_lambda)



# 3. dictionary comprehension

dict_comp = {key:value for key, value in d.items() if value >=25} # key:value는 리스트 컴프리헨션의 딕셔너리 버전 (dict comprehension)을 만드는 공식 
print(dict_comp)
```

```python
# 3개 리스트를 {'key': a, 'value': b*c} 딕셔너리 구조로 변경

a = ['one', 'two', 'three', 'four']
b= [30,20,15,75]
c = [5.2, 7.4, 3.6, 4,2]

# 방법 1
emp = {}
for x,y, z in zip(a,b,c): # zip(,): 복수의 리스트를 동시에 이터레이트 시키도록 묶어줌
    val = y*z
    emp[x] = val # 리스트 append의 딕셔너리 버전. 딕셔너리는 append 없음
print(emp)
    
    
# 방법 2 - for문 썼다는 것은 리스트 컴프리헨션도 가능하다는 것
{x: y * z for x,y,z in zip(a,b,c)}
# 리스트/딕셔너리 컴프리헨션은 빈 리스트/딕셔너리 생성 추기 작업을 자동으로 해줌. 내가 안해도 됨. 최종 결과가 무엇인지만 넣으면 됨


# 방법 3
dict([x, y*z] for x,y,z in zip(a,b,c))
# dict()는 쌍으로 표현되는 인자를 받아야함 (길이 2개 튜플 or 리스트) -> 그래서 x, y*z 감싼 것이 []
```

```python
# split lists into N chunks
# N에 대한 input을 받아 유연하게 만들기. input -> 함수 만들기

x = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O', 'P', 'Q']

# 방법 1

# n= 3

def split_list (n):
    emp = []

    for i in range(0, len(x), n): # 0,3,6,...
        # range 속 step 기능 잊지 말고 활용!
        emp.append(x[i:i+n])
    print(emp)
    
split_list(3)

    
# 방법 2: list comprehension - for 쓸 수 있으니 리스트 컴프리헨션도 가능

def split_list_comp(n):
    print([x[i:i+n] for i in range (0, len(x), n)])
    
split_list_comp(3)
```

```python
# 사용자에게 정수를 3회 입력받은 뒤 평균값으로 출력. 

def avg_func(a,b,c):
    for i in [a,b,c]:
        emp = []
        emp.append(i)
    return sum(emp)/len(emp)
    


avg_func(1,2,3)    
```

```python
word = 'secrect' # 정답 단어
guesses = '' # 추측 단어
turns = 10 # 기회 10번

# 기회 1번 남을 때까지 계속 함
while turns >0:
    failed = 0 # 실패 횟수
    
    # 정답 맞추게 제시.
    # 단어에 있으면 출력, 없으면 failed 에 추가
    for char in word:
        if char in guesses:
            print(char, end = ' ')
        else:
            print("_", end = ' ')
            failed += 1
            
    # 단어 다 맞췃으면 while 구문 중단  
    if failed == 0:
        print("Correct")
        break
    guess = input("your turn")
    guesses += guess
    
    # 정답 단어에 추측 문자 없으면
    if guess not in word:
        turns -= 1
        print("Wrong. You have", turns, 'more guess')
        
    # 기회 다 쓰면 게임 오버
    if turns == 0:
        print('Game over')
```
