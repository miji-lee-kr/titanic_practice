```python
# value 값이 25 이상인 Dict 요소만 골라내자 - 복습

d = {'a': 8, 'b':33, 'c': 3}

# 1. 일반 루프

emp = {}

for key, value in d.items(): # 그냥 for k,v in d: 로 하면 안됨.
                        # d는 dict 전체를 받았어도 루프를 도는건 key뿐이다. value는 print(value) 처럼 아래에서 따로 부르지 않는 한 숨어있음
                        # d.items()는 (a,8) 같이 key, value 모두 있는 하나의 튜플을 받고, 튜플 각 쌍이 iterate한다.
                        # 즉 둘 다 필요하다면 for in d.items()를 써야함. value만 필요하면 for in d.value()를 받자.
                        # 한번 돌아서 받은 (a, 8) 튜플은 for구문이 자동으로 언패킹해서 a는 key에, 8은 value에 자동으로 어사인해준다 = 언패킹.
                        # d는 key라는 하나의 변수밖에 못받으므로 key, value 두개 요청한 갯수랑 안맞아 에러 
    if value >=25:
        emp[key] = value
        # emp에 append 안하고 if value >=25: print(d) 하니 모든 딕셔너리가 나왔던 실수 -
        # print()에 전달된건 여전히 d 전체이므로, value >=25 인 값이 하나라도 있으면 전체를 표시하라는 지시가 됨
        # 차라리 print(key,value)하면 해당되는 부분만 나옴
print(emp)


# 2. lambda

dict_lambda = dict(filter(lambda x: x[1] >=25, d.items())) # d만 쓰면 lambda도 for처럼 key만 받음. 람다에서도 d.items() 를 받아야 
                                                            # d.items() 로 한번 루프 돌 때마다 (a,8) 같이 두개 변수를 가진 한 덩이의 튜플 생김
                                                            # lambda는 하나의 변수만 받아줌 & for처럼 key, value for d.items() 같은 자동 언팩킹 안해줌 -> key, value에 자동으로 값 assign 안해줌
                                                            # 우리가 튜플 안에서 뭐가 value인지 수동으로 assign 해줘야함 -> (a,8)에서 1번째 인덱스가 value = tuple[1]
                                                            # key>=25인 튜플 덩어리만 필터링해서 그 덩어리를 dict로 다시 패킹
print(dict_lambda)



# 3. dictionary comprehension

dict_comp = {key:value for key, value in d.items() if value >=25} # key:value는 리스트 컴프리헨션의 딕셔너리 버전 (dict comprehension)을 만드는 공식 
print(dict_comp)
```

```python
# split lists into N chunks
# N에 대한 input을 받아 유연하게 만들기. input -> 함수 만들기

x = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O', 'P', 'Q']

# 방법 1

# n= 3

def split_list (n):
    emp = []

    for i in range(0, len(x), n): # 0,3,6,...
        # range 속 step 기능 잊지 말고 활용!
        emp.append(x[i:i+n])
    print(emp)
    
split_list(3)

    
# 방법 2: list comprehension - for 쓸 수 있으니 리스트 컴프리헨션도 가능

def split_list_comp(n):
    print([x[i:i+n] for i in range (0, len(x), n)])
    
split_list_comp(3)
```

```python
# 쿠폰 코드 생성
# 6자리 무작위 코드를 중복없이 N개 생성해 한 리스트에 담기

import random

char = 'abcdefghijklmnopqrstuvwxyz0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ!@#$%^&*()?'

# 방법 1

def generate (num):
    emp_list = []
    for i in range(num):
        rand = random.sample(char, k=6)
        emp_list.append("".join(rand))    
    return emp_list

generate(3)


# 방법 2 : set() 집합 활용

def generate(num):
    emp_set = set() # 빈 set 정의
    while len(emp_set) < num:
        rand = random.choices(char, k= 6)
        emp_set.add(''.join(rand)) # append의 set 버전
    return emp_set

generate(4)
```

```python
# 행맨 게임 만들기

word = 'secrect' # 정답 단어
guesses = '' # 추측 단어
turns = 10 # 기회 10번

# 기회 1번 남을 때까지 계속 함
while turns >0:
    failed = 0 # 실패 횟수
    
    # 정답 맞추게 제시.
    # 단어에 있으면 출력, 없으면 failed 에 추가
    for char in word:
        if char in guesses:
            print(char, end = ' ')
        else:
            print("_", end = ' ')
            failed += 1
            
    # 단어 다 맞췃으면 while 구문 중단  
    if failed == 0:
        print("Correct")
        break
    guess = input("your turn")
    guesses += guess
    
    # 정답 단어에 추측 문자 없으면
    if guess not in word:
        turns -= 1
        print("Wrong. You have", turns, 'more guess')
        
    # 기회 다 쓰면 게임 오버
    if turns == 0:
        print('Game over')
```

```python
# 전체 제품 가격 인상하는 코드

# 방법 1. 인스턴트 메소드로만

class Car():
    
    def __init__ (self, brand, details): 
        self._brand = brand 
        self._details = details


# 인스턴스 메소드로 개별 price가 무엇에 의해 결정되는지 바꿔주기
    def price_updated (self, pct):
        self._details['price'] *= pct
    
    
car1 = Car('Ferrari', {'colour':'white', 'price': 4000})
car2 = Car('BMW', {'colour':'silver', 'price': 3000})
car3 = Car('Audi', {'colour':'black', 'price': 2000})


# 구체적인 숫자 넣어 본격적으로 적용    
for c in [car1, car2, car3]:
    c.price_updated(1.3)
    print(f'{c._brand}, {c._details['price']} EUR')
```

```python
# 방법 2. 클래스 메소드 포함해서

# 가격 올릴 때 항상 변경하는 클래스 변수 increase_pct 가 있다고 하자
# !!! 놓친 부분!!! increase_pct를 내가 원하는 대로 먼저 바꿔야. 다리 역할. 클래스 인스턴스를 참조하니 클래스 메소드
# 그럼 가격이 그걸 반영해야 되자나 : price *= increase_pct !!! 놓친 부분: 각 객체가 자기 가격을 변경하니 인스턴스 메소드

class Car():
    
    increase_pct = 1.0
    
    def __init__ (self, brand, details): 
        self._brand = brand 
        self._details = details
        
    @classmethod
    def pct_change (cls, new_pct):
        cls.increase_pct = new_pct 
        
    def updated_price (self):
        self._details['price'] *= Car.increase_pct
        return self._details['price']
        
    
car1 = Car('Ferrari', {'colour':'white', 'price': 4000})

car1.pct_change(1.5) # 퍼센티지 바꾸는 함수도 함께 호출 잊지 말기
print(Car.updated_price(car1))
```
