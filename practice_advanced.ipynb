```python
# í´ë˜ìŠ¤ ë©”ì†Œë“œ

class Car():

    increase_pct = 1.2

    def __init__ (self, brand, details): 
        self._brand = brand 
        self._details = details
    
    @classmethod
    def price_change (cls, pct): # pctëŠ” ì—¬ê¸°ì„œ ì •ì˜ë˜ì§€ ì•Šê³ , í˜¸ì¶œí•  ë•Œ ì œì‹œë˜ëŠ” ìˆ«ì
        cls.increase_pct = pct

    def after_price(self):
        return f'Updated price : {self._brand}, {self._details.get('price')* Car.increase_pct} EUR'


car1 = Car('Ferrari', {'colour':'white', 'price': 4000})

car1.price_change(1.3)
print(car1.after_price())
```

```python
# special method : Fruit.eat(f1) ì²˜ëŸ¼ ë‚´ê°€ "ì´ë¦„ìœ¼ë¡œ" ìˆ˜ë™ í˜¸ì¶œí•´ì•¼ ì‹¤í–‰ë˜ëŠ” ë©”ì„œë“œ ë§ê³ , "íŠ¹ì • ë¬¸ë²• ì‹¤í–‰ì‹œ" ìë™ í˜¸ì¶œë˜ëŠ” ë©”ì„œë“œ
# __str__()ê°™ì€ ëª¨ì–‘. ë‚´ê°€ ë§Œë“œëŠ” ì¼ë°˜ ë©”ì„œë“œëŠ” ì—°ê²°ëœ ë¬¸ë²•ì´ ì—†ìœ¼ë¯€ë¡œ ìˆ˜ë™ í˜¸ì¶œí•´ì•¼

class Fruit:
        
    def __init__(self, name, price):
        self._name = name
        self._price = price
        
    def __repr__(self):
        return f'Class info {self._name}, {self._price}'
        
    def __sub__(self, other):
        return (other._price * self._price) / 3
    
    def eat (self):
        return print(self._name, "is yummy")

f1 = Fruit('orange', 100)
f2 = Fruit('banana', 500)

results = f2-f1
print(results) # ë”°ë¡œ Fruit.__sub__(f1,f2,f3) íŠ¹ì • ë©”ì†Œë“œë¡œ ì•ˆë¶ˆëŸ¬ë„ - ëŠ” __sub__ê°€ ìë™ ì‹¤í–‰ë˜ê²Œ ëª…ë ¹ ë‚´ë¦¼.
# ì—°ì‚° ë¡œì§ ì»¤ìŠ¤í„°ë§ˆì´ì¦ˆí•´ì„œ ìƒˆ ê¸°ëŠ¥ ë§Œë“¤ê¸° ê°€ëŠ¥. ë¹¼ê¸° ê¸°í˜¸ë¥¼ ë„£ì—ˆëŠ”ë° ë‚´ë¶€ì ìœ¼ë¡œëŠ” ê³±í•˜ê¸° ë˜ë„ë¡ í•˜ëŠ” ê²ƒì²˜ëŸ¼

reaction = Fruit.eat(f1)
```
