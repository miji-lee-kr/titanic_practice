```python
word = 'secrect' # 정답 단어
guesses = '' # 추측 단어
turns = 10 # 기회 10번

# 기회 1번 남을 때까지 계속 함
while turns >0:
    failed = 0 # 실패 횟수
    
    # 정답 맞추게 제시.
    # 단어에 있으면 출력, 없으면 failed 에 추가
    for char in word:
        if char in guesses:
            print(char, end = ' ')
        else:
            print("_", end = ' ')
            failed += 1
            
    # 단어 다 맞췃으면 while 구문 중단  
    if failed == 0:
        print("Correct")
        break
    guess = input("your turn")
    guesses += guess
    
    # 정답 단어에 추측 문자 없으면
    if guess not in word:
        turns -= 1
        print("Wrong. You have", turns, 'more guess')
        
    # 기회 다 쓰면 게임 오버
    if turns == 0:
        print('Game over')
```
